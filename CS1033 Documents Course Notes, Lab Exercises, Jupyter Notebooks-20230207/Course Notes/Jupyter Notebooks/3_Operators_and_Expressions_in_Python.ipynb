# 3.1  Types of Operators

In the expression 4 + 5, 4 and 5 are operands and + is the operator. Python language supports the following types of operators.


*   Arithmetic operators
*   Comparison (i.e., relational) operators
*   Assignment operators
*   Bitwise operators
*   Logical operators
*   Membership operators
*   Identity operators

Let's look at these types one by one.

# 3.5  Python Bitwise  Operators

Bitwise operators work on bits; that is, they perform a bit-by-bit operation on operands. Assume a = 60 and b = 13. The following show their binary representation and 4 bitwise operations on them.

	a 	=   0011 	1100
	b	 =   0000 	1101
	--------------------------
	a&b = 	0000 	1100
	a|b = 	0011 	1101
	a^b = 	0011 	0001
	~a  = 	1100 	0011

The following bitwise operators are supported by Python.

<table align="center">
  <tr>
      <th style="text-align:center">Operator</th>
      <th style="text-align:left">Description</th>
      <th style="text-align:left">Example</th>
  </tr>
  
  <tr>
      <td align="center" style="text-align:center">&</td>
      <td align="center" style="text-align:left"> Binary AND Operator: sets a bit to 1 in the result if the corresponding bit is 1 in both operands. </td>
      <td align="center" style="text-align:left"> (a & b) will give 12 which is 0000 1100</td>
  </tr>
  <tr>
      <td align="center" style="text-align:center">|</td>
      <td align="center" style="text-align:left"> Binary OR Operator: sets a bit to 1 in the result if the corresponding bit is 1 in either operands.</td> 
      <td align="center" style="text-align:left"> (a | b) will give 61 which is 0011 1101</td>
  </tr>
  <tr>
      <td align="center" style="text-align:center">^</td>
      <td align="center" style="text-align:left"> Binary XOR Operator: sets a bit to 1 in the result if the corresponding bit is 1 in exactly one operand. </td>
      <td align="center" style="text-align:left"> (a ^ b) will give 49 which is 0011 0001 </td>
  </tr>
  <tr>
      <td align="center" style="text-align:center">~</td>
      <td align="center" style="text-align:left"> Binary Ones Complement Operator: is unary and has the effect of 'flipping' bits. </td>
      <td align="center" style="text-align:left"> (~a ) will give -61 which is 1100 0011 in 2's complement</td>
  </tr>
  <tr>
      <td align="center" style="text-align:center"><<</td>
      <td align="center" style="text-align:left"> Binary Left Shift Operator: The left operand’s value is moved left by the number of bits specified by the right operand.</td> 
      <td align="center" style="text-align:left">a << 2 will give 240 which is 1111 0000</td>
  </tr>
  <tr>
      <td align="center" style="text-align:center">>></td>
      <td align="center" style="text-align:left">Binary Right Shift Operator: The left operand’s value is moved right by the number of bits specified by the right operand. </td>
      <td align="center" style="text-align:left"> a >> 2 will give 15 which is 0000 1111</td>
  </tr>
</table>
